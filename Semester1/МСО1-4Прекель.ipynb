### Задача 8

Решение:

Событие А – наудачу по одному извлекли нумерованные кубики, номера извлеченных кубиков появились в возрастающем порядке.

Общее число равновозможных исходов доставания 6 кубиков равно $P_6 = $


```python
def fact(a):
    if a == 0:
        return 1
    else:
        return a * fact(a - 1)
    
total = fact(6)
total
```

Исход, благоприятствующие событию $А$ всего один: $(1,2,3,4,5,6)$, тогда $P(A) = $

```python
Fraction(1, total)
```

### Задача 9

Решение:

Событие А – бросили три кости, на одной из них выпала 6, на двух других выпали числа, не совпадающие между собой и не равные 6.

Общее число исходов бросания костей, когда все значения разные, равно 


```python
def C(k, n):
    return fact(n) // (fact(k) * fact(n - k))

total = C(3, 6)
total
```

Число исходов, благоприятствующих событию А, то есть, когда одна кость равна 6, а остальные любые не равные друг другу и не равные 6, равно числу сочетаний из 5 элементов по 2:

```python
a = C(2, 5)
a
```

тогда $P(A) = $

```python
Fraction(a, total)
```

Ответ: $\frac{1}{2}$
