## –í–∞—Ä–∏–∞–Ω—Ç 1

### –ó–∞–¥–∞–Ω–∏–µ 1

> –≠–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫–∞—è  —Ü–µ–ø—å  —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–∞  –ø–æ  —Å—Ö–µ–º–µ,  –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω–æ–π  –Ω–∞ —Ä–∏—Å—É–Ω–∫–µ.  –°–æ–±—ã—Ç–∏–µ ùê¥ùëò  –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è  –≤  —Ç–æ–º,  —á—Ç–æ ùëò-—ã–π  —ç–ª–µ–º–µ–Ω—Ç  –∏—Å–ø—Ä–∞–≤–µ–Ω. –ó–∞–ø–∏—Å–∞—Ç—å —Å–æ–±—ã—Ç–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ –≤—Å—è —Å–∏—Å—Ç–µ–º–∞ –∏—Å–ø—Ä–∞–≤–Ω–∞ ùê¥, —Å–∏—Å—Ç–µ–º–∞ –Ω–µ–∏—Å–ø—Ä–∞–≤–Ω–∞ ùêµ.

$A_i$ - $i$-—ã–π —ç–ª–µ–º–µ–Ω—Ç –∏—Å–ø—Ä–∞–≤–µ–Ω

$A = (A_1 + A_2)(A_3 + A_4)$

$B = \not{A} = \not{A_1}\not{A_2} + \not{A_3}\not{A_4}$

–û—Ç–≤–µ—Ç: $A = (A_1 + A_2)(A_3 + A_4)$; $B = \not{A_1}\not{A_2} + \not{A_3}\not{A_4}$

### –ó–∞–¥–∞–Ω–∏–µ 2

> –ù–∞–ø–∏—à–∏—Ç–µ —Ñ–æ—Ä–º—É–ª—É —á–∏—Å–ª–∞ —Å–æ—á–µ—Ç–∞–Ω–∏–π

```python
def fact(a):
    if a == 0:
        return 1
    else:
        return a * fact(a - 1)
    
def C(k, n):
    return fact(n) // (fact(k) * fact(n - k))
```

–û—Ç–≤–µ—Ç: $–°_n^k = \frac{n!}{k!(n - k)!}$

### –ó–∞–¥–∞–Ω–∏–µ 3

> –í –≥—Ä—É–ø–ø–µ –æ–±—Å–ª–µ–¥—É–µ–º—ã—Ö 1000 —á–µ–ª–æ–≤–µ–∫. –ò–∑ –Ω–∏—Ö 600 –∫—É—Ä—è—â–∏—Ö –∏ 400 –Ω–µ–∫—É—Ä—è—â–∏—Ö. –°—Ä–µ–¥–∏ –∫—É—Ä—è—â–∏—Ö 240 —á–µ–ª–æ–≤–µ–∫ –∏–º–µ—é—Ç —Ç–µ –∏–ª–∏ –∏–Ω—ã–µ –∑–∞–±–æ–ª–µ–≤–∞–Ω–∏—è –ª–µ–≥–∫–∏—Ö.  –°—Ä–µ–¥–∏  –Ω–µ–∫—É—Ä—è—â–∏—Ö  –ª–µ–≥–æ—á–Ω—ã—Ö  –±–æ–ª—å–Ω—ã—Ö  120  —á–µ–ª–æ–≤–µ–∫.  –Ø–≤–ª—è—é—Ç—Å—è  –ª–∏ –∫—É—Ä–µ–Ω–∏–µ –∏ –∑–∞–±–æ–ª–µ–≤–∞–Ω–∏–µ –ª–µ–≥–∫–∏—Ö –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏ —Å–æ–±—ã—Ç–∏—è–º–∏?

–ü—É—Å—Ç—å A - –æ–±—Å–ª–µ–¥—É–µ–º—ã–π –∫—É—Ä–∏—Ç, B - –æ–±—Å–ª–µ–¥—É–µ–º—ã–π —Å –∑–∞–±–æ–ª–µ–∞–Ω–∏–µ–º –ª—ë–≥–∫–∏—Ö.

–¢–æ–≥–¥–∞ $P(B) = $

```python
from fractions import Fraction

PB = Fraction(240 + 120, 1000)
PB
```

$P_A(B) = $

```python
PAB = Fraction(240, 600)
PAB
```

–¢–∞–∫ –∫–∞–∫ $P(B) \neq P_A(B)$, —Å–æ–±—ã—Ç–∏—è A –∏ B –∑–∞–≤–∏—Å–∏–º—ã–µ.

### –ó–∞–¥–∞–Ω–∏–µ 4

> –ò–º–µ—é—Ç—Å—è —Ç—Ä–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö –ø–æ –≤–∏–¥—É —è—â–∏–∫–∞. –í –ø–µ—Ä–≤–æ–º –Ω–∞—Ö–æ–¥—è—Ç—Å—è –¥–≤–µ –±–µ–ª—ã–µ –º—ã—à–∏ –∏ –æ–¥–Ω–∞ —Å–µ—Ä–∞—è, –≤–æ –≤—Ç–æ—Ä–æ–º ‚Äì —Ç—Ä–∏ –±–µ–ª—ã–µ –∏ –æ–¥–Ω–∞ —Å–µ—Ä–∞—è, –≤ —Ç—Ä–µ—Ç—å–µ–º ‚Äì–¥–≤–µ  –±–µ–ª—ã–µ  –∏  –¥–≤–µ  —Å–µ—Ä—ã–µ  –º—ã—à–∏.  –ö–∞–∫–æ–≤–∞  –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å  —Ç–æ–≥–æ,  —á—Ç–æ  –∏–∑  –Ω–∞—É–≥–∞–¥ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ —è—â–∏–∫–∞ –±—É–¥–µ—Ç –∏–∑–≤–ª–µ—á–µ–Ω–∞ –±–µ–ª–∞—è –º—ã—à—å?

–°–æ–±—ã—Ç–∏–µ $–ê$ - –∏–∑–≤–ª–µ—á–µ–Ω–∞ –±–µ–ª–∞—è –º—ã—à—å. –ì–∏–ø–æ—Ç–µ–∑—ã: $–í_1$ - –∏–∑ –ø–µ—Ä–≤–æ–≥–æ —è—â–∏–∫–∞; $–í_2$ - –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ; $–í_3$ - –∏–∑ —Ç—Ä–µ—Ç—å–µ–≥–æ.

–¢–æ–≥–¥–∞ $–†(–í_1)=–†(–í_2)=–†(–í_3)=\frac{1}{3}$

$P_{B_1}(A)=\frac{2}{3}$

$P_{B_2}(A)=\frac{3}{4}$

$P_{B_3}(A)=\frac{2}{4}$

```python
PB123 = Fraction(1, 3)
PB1A = Fraction(2, 3)
PB2A = Fraction(3, 4)
PB3A = Fraction(2, 4)

PB123, PB1A, PB2A, PB3A
```

–ü–æ —Ñ–æ—Ä–º—É–ª–µ –ø–æ–ª–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ $–†(–ê) = P(B_1) * P_{B_1}(A) + P(B_2) * P_{B_2}(A) + P(B_3) * P_{B_3}(A) \approx$

```python
PA = PB123 * PB1A + PB123 * PB2A + PB123 * PB3A;
PA, round(float(PA), 2)
```

–û—Ç–≤–µ—Ç: $0.64$

### –ó–∞–¥–∞–Ω–∏–µ 5

> –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—ã–ø–∞–¥–µ–Ω–∏—è  —ç–ø–∏—á–Ω–æ–≥–æ –ª—É—Ç–∞ –∏–∑ –ª—É—Ç–±–æ–∫—Å–∞ —Ä–∞–≤–Ω–∞ 0.05. –ò–≥—Ä–æ–∫ –æ—Ç–∫—Ä—ã–≤–∞–µ—Ç 100 –ª—É—Ç–±–æ–∫—Å–æ–≤. –ù–∞–π—Ç–∏    –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —ç–ø–∏—á–µ—Å–∫–∞—è –≤–µ —â –≤—ã–ø–∞–¥–µ—Ç    —Ä–æ–≤–Ω–æ 2 —Ä–∞–∑–∞.

–ü–æ —Ñ–æ—Ä–º—É–ª–µ –ë–µ—Ä–Ω—É–ª–ª–∏: $P(A) = C_{100}^2 * p^2 * (1 - p)^{98} \approx$

```python
p = 0.05
PA = C(2, 100) * p**2 * (1 - p) ** 98
PA, round(PA, 4)
```

–ü—Ä–æ–≤–µ—Ä–∏–º —Å –ø–æ–º–æ—â—å—é –æ–ø—ã—Ç–∞ –∏—Å–ø–æ–ª—å–∑—É—è –≥–µ–Ω–µ—Ä–∞—Ü–∏—é —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª:

```python
from random import randint
from fractions import Fraction

def random_epic():
    if randint(1, 100) <= 5:
        return True
    else:
        return False

def open_100():
    c = 0
    for i in range(0, 100):
        if random_epic():
            c += 1
    return c

def calcP(tries):
    r = 0
    for i in range(tries):
        if open_100() == 2:
            r += 1
    return Fraction(r, tries)

for i in range(0, 9):
    tries = 2 ** i * 100
    print(str(tries) + " –æ–ø—ã—Ç–æ–≤: " + str(float(calcP(tries))))
```

–í–∏–¥–Ω–æ, —á—Ç–æ –ø—Ä–∏ –±–æ–ª—å—à–æ–º –∫–æ–ª-–≤–æ –æ–ø—ã—Ç–æ–≤ –æ—Ç–≤–µ—Ç —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –≤—ã—á–∏—Å–ª–µ–Ω–æ–º—É –ø–æ —Ñ–æ—Ä–º—É–ª–µ –ë–µ—Ä–Ω—É–ª–ª–∏.

–û—Ç–≤–µ—Ç: $0.0812$
