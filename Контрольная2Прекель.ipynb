### Задание 1

```python
import sympy as sp
```

```python
x, y, a = sp.symbols("x y a")
x, y, a
```

```python
f_x = a * (1 - x**2)
f_x
```

Найдём параметр $a$

```python
f_x_int = sp.integrate(f_x, (x, -1, 1))
f_x_int
```

```python
a = sp.solve(sp.Eq(f_x_int, 1))[0]
a
```

Найдём мат. ожидание

```python
f_x = a * (1 - x ** 2)
f_x
```

```python
m_x = sp.integrate(x * f_x, (x, -1, 1))
m_x
```

Найдём дисперсию

```python
d_x = sp.integrate(f_x * (x - m_x) ** 2, (x, -1, 1))
d_x
```

Найдем среднеквадратическое отклонение

```python
s = d_x ** (1/2)
s
```

Медиана $F(x) = \frac{1}{2}$:

```python
sp.solve(sp.Eq(sp.integrate(f_x, (x, -1, x)), 1/2), x)
```

### Задание 3

```python
xi = [-1, 0, 1, 2]
xi
```

```python
u = sp.symbols("u")
pi = [0.1, 0.4, u, 0.4]
pi
```

```python
pi[2] = sp.solve(sum(pi) - 1)[0]
pi
```

Производящая функция:

$Q(z) = $

```python
z = sp.symbols("z")
Q = 0
for i in range(0, 4):
    Q += pi[i] * z**i
Q
```

Математическое ожидание:

```python
m = 0
for i in range(len(xi)):
    m += xi[i] * pi[i]
m
```

Дисперсия:

```python
d = -0.8**2
for i in range(len(xi)):
    d += xi[i] ** 2 * pi[i]
d
```

Среднеквадратичное отклонение:

```python
from math import sqrt
sigma = sqrt(d)
sigma
```

Медиана:

```python
for i in range(0, len(xi) + 1):
    print(f"F({i}) = ", end='')
    F = 0
    for j in range(i):
        F += pi[j]
    print(F)
```
